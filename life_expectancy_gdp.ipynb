```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy as sp
import statsmodels.api as sm
```

```python
life_expectancy_df = pd.read_csv('all_data.csv')
print(life_expectancy_df.head())
print(life_expectancy_df.dtypes)
print(life_expectancy_df.columns)
print(life_expectancy_df.isna().sum())
print(life_expectancy_df.describe())
print(life_expectancy_df.Country.unique())
years = range(2000,2016)
```

```python
plt.figure(figsize=(8,10))
ax = plt.subplot()
plt.plot(range(16), life_expectancy_df.loc[life_expectancy_df.Country == 'Chile'].GDP, label = 'Chile', color = '#004949', marker = 'o')

plt.plot(range(16), life_expectancy_df.loc[life_expectancy_df.Country == 'China'].GDP, label = 'China', color = '#ff6db6', marker = 'o')

plt.plot(range(16), life_expectancy_df.loc[life_expectancy_df.Country == 'Germany'].GDP, label = 'Germany', color = '#490092', marker = 'o')

plt.plot(range(16), life_expectancy_df.loc[life_expectancy_df.Country == 'Mexico'].GDP, label = 'Mexico', color = '#b66dff', marker = 'o')

plt.plot(range(16), life_expectancy_df.loc[life_expectancy_df.Country == 'United States of America'].GDP, label = 'United States', color = '#b6dbff', marker = 'o')

plt.plot(range(16), life_expectancy_df.loc[life_expectancy_df.Country == 'Zimbabwe'].GDP, label = 'Zimbabwe', color = '#7DF95F', marker = 'o')

ax.set_xticks(range(16))
ax.set_xticklabels(years, rotation = 45)
highest_GDP = life_expectancy_df.GDP.max()
yticks = []
for i in range(10):
    yticks.append(i * highest_GDP/10)

ax.set_yticks(yticks)

plt.ylim(bottom = 0)
plt.xlim(0, 15)

plt.xlabel('Year')
plt.ylabel('GDP in Ten Trillions')
plt.legend(loc = 'upper left')
plt.title('GDP of Countries over 15 Years')

plt.show()
plt.clf()
```

```python
plt.figure(figsize=(12,12))
yticks_le = range(70, 90, 5)
ax1 = plt.subplot(3,2,1)
plt.subplot(3,2,1)
plt.plot(range(2000,2016), life_expectancy_df.loc[life_expectancy_df.Country == 'Chile']['Life expectancy at birth (years)'], color = '#004949')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
ax1.set_xticks(range(2000,2016))
ax1.set_xticklabels(range(2000,2016), rotation = 45)
ax1.set_yticks(yticks_le)
plt.title('Life Expectancy Over Time in Chile')

ax2 = plt.subplot(3,2,2)
plt.subplot(3,2,2)
plt.plot(range(2000,2016), life_expectancy_df.loc[life_expectancy_df.Country == 'China']['Life expectancy at birth (years)'], color = '#ff6db6')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
ax2.set_yticks(yticks_le)
ax2.set_xticks(range(2000,2016))
ax2.set_xticklabels(range(2000,2016), rotation = 45)
plt.title('Life Expectancy Over Time in China')

ax3 = plt.subplot(3,2,3)
plt.subplot(3,2,3)
plt.plot(range(2000,2016), life_expectancy_df.loc[life_expectancy_df.Country == 'Germany']['Life expectancy at birth (years)'], color = '#490092')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
ax3.set_yticks(yticks_le)
ax3.set_xticks(range(2000,2016))
ax3.set_xticklabels(range(2000,2016), rotation = 45)
plt.title('Life Expectancy Over Time in Germany')

ax4 = plt.subplot(3,2,4)
plt.subplot(3,2,4)
plt.plot(range(2000,2016), life_expectancy_df.loc[life_expectancy_df.Country == 'Mexico']['Life expectancy at birth (years)'], color = '#b66dff')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
ax4.set_yticks(yticks_le)
ax4.set_xticks(range(2000,2016))
ax4.set_xticklabels(range(2000,2016), rotation = 45)
plt.title('Life Expectancy Over Time in Mexico')

ax5 = plt.subplot(3,2,5)
plt.subplot(3,2,5)
plt.plot(range(2000,2016), life_expectancy_df.loc[life_expectancy_df.Country == 'United States of America']['Life expectancy at birth (years)'], color = '#b6dbff')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
ax5.set_yticks(yticks_le)
ax5.set_xticks(range(2000,2016))
ax5.set_xticklabels(range(2000,2016), rotation = 45)
plt.title('Life Expectancy Over Time in the United States')

ax6 = plt.subplot(3,2,6)
plt.subplot(3,2,6)
plt.plot(range(2000,2016), life_expectancy_df.loc[life_expectancy_df.Country == 'Zimbabwe']['Life expectancy at birth (years)'], color = '#7DF95F')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
ax6.set_yticks([40,45,50,55,60])
ax6.set_xticks(range(2000,2016))
ax6.set_xticklabels(range(2000,2016), rotation = 45)
plt.title('Life Expectancy Over Time in Zimbabwe')


plt.subplots_adjust(hspace=.5)
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(8,10))
plt.hist(life_expectancy_df.loc[life_expectancy_df.GDP > life_expectancy_df.GDP.median()]['Life expectancy at birth (years)'], density = True, bins = 10, color = 'blue', alpha = .3, label = 'Life Expectancy of Countries Above the Median GDP')
plt.hist(life_expectancy_df.loc[life_expectancy_df.GDP < life_expectancy_df.GDP.median()]['Life expectancy at birth (years)'], density = True, bins = 20, color = 'orange', alpha = .3, label = 'Life Expectancy of Coutries Under the Median GDP')
plt.legend(loc= 'upper left')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('Proportion')
plt.title('Life Expectancy by Relation to Median GDP')
plt.show()
plt.clf()
```

```python
countries = ['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe']
colors = ['blue', 'green', 'red', 'orange', 'purple', 'cyan']
plt.figure(figsize = (8,10))
for index, country in enumerate(countries):
    GDP_min = int(life_expectancy_df.loc[life_expectancy_df.Country == country].GDP.min())
    GDP_max = int(life_expectancy_df.loc[life_expectancy_df.Country == country].GDP.max())
    ax = plt.subplot(3,2,index+1)
    plt.subplot(3,2,index+1)
    plt.scatter(life_expectancy_df.loc[life_expectancy_df.Country == country].GDP, life_expectancy_df.loc[life_expectancy_df.Country == country]['Life expectancy at birth (years)'], color = colors[index], label = country)
    plt.xlabel('GDP in Trillions')
    plt.ylabel('Life Expectancy in Years')
    plt.title('GDP vs Life Expectancy in ' + country)
    ax.set_xticks([GDP_min, GDP_max])
    ax.set_xticklabels(np.round([GDP_min/10e11, GDP_max/10e11],3))

plt.subplots_adjust(hspace=.8, wspace = .7)
plt.show()
plt.clf()
```
